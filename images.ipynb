```python
from diffusers import StableDiffusionPipeline
import torch

# Проверяем, есть ли доступная GPU
device = "cuda" if torch.cuda.is_available() else "cpu"

# Загружаем модель Stable Diffusion
pipeline = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    revision="fp16",
    torch_dtype=torch.float16 if device == "cuda" else torch.float32,  # Меняем тип данных в зависимости от устройства
)
pipeline.to(device)

print("Модель успешно загружена и готова к использованию!")
```

```python
# Указываем описание для генерации изображения
prompt = "cartoon squirrel"

# Генерируем изображение
image = pipeline(prompt).images[0]

# Показываем изображение
image.show()

# Сохраняем изображение на диск
image.save("generated_image.png")

print("Изображение сохранено как 'generated_image.png'")
```
